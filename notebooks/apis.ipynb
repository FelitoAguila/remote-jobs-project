### Testing Jobicy api

```python
import requests

url = 'https://jobicy.com/api/v2/remote-jobs'
params = {
    'count': 10,
    'tag': 'data analyst'
}
response = requests.get(url, params=params)
data = response.json()
jobs = data.get('jobs', [])
```

### Testing duckdb

```python
import duckdb

conn = duckdb.connect(database='remote_jobs.duckdb')

# Set search path to the dataset
conn.sql(f"SET search_path = 'remote_jobs_data'")

# Describe the dataset
conn.sql("DESCRIBE")
```

```python
remote_jobs = conn.sql("SELECT * FROM remote_jobs").df()
dlt_loads = conn.sql("SELECT * FROM remote_jobs__job_type      ").df()
```

```python
# Set search path to the dataset
conn.sql(f"SET search_path = 'remote_jobs_data_staging'")
remote_jobs_staging = conn.sql("SELECT * FROM remote_jobs").df()
```
